```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
college_data = pd.read_csv('experiment-6.csv')
```

```python
college_data.head(10)
```

```python
college_type = []
slope = []

for _, row in college_data.iterrows():
    if not pd.isna(row['Slope']):
        college_type.append(row['Type'])
        slope.append(row['Slope'])

print(college_type)
print(slope)
```

```python
df = pd.DataFrame({
    'Type': college_type,
    'Slope': slope
})

df['Type'] = pd.Categorical(df['Type'], categories=['ivy-league', 'liberal-arts', 'research'])
df
```

```python
from plotnine import *
```

```python
plot = (ggplot(df) +
  geom_boxplot(aes(y = 'slope', x = 'factor(Type)', fill = 'Type')) +
  coord_flip() +
  ylab("Bipartisan Slope") +
  xlab("College Types") +
  labs(fill="College Type") +
  ggtitle("Bipartisan Trends Across Different Institutional Types")
)
```

```python
plot
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('Slope ~ C(Type)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```
